# PCA Correlation Circle for Gas Leaks Data

```python
from mlxtend.data import iris_data
from mlxtend.plotting import plot_pca_correlation_graph
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'https://raw.githubusercontent.com/MarinaOrzechowski/GasLeakConEd/timeline_branch/data/processed/important_(used_in_app)/ready_for_PCA.csv')
df = df.drop(columns ={'Unnamed: 0', 'geoid','gentrified', 'gas_leaks'})
```

```python
df = df[['avg_bldg_age', '311total_compl_pp', '%311: water system',
       '%311: noise', '%311: plumbing', '%311: illegal parking',
       '%311: unsanitary condition', '%311: sewer',
       '%311: street light condition', '%311: heating',
       '%311: street condition', '%311: blocked driveway',
       '%311: sanitation condition', '%311: paint/plaster',
       '%311: dirty conditions', '%311: request large bulky item collection',
       '%311: heat/hot water', '%housh_grandp_resp_for_grandch',
       'avg_houshold_size',  'lonely_housholder%',
       'lonely_housholder_over65%', 'mean_houshold_income',
       'nonfamily_housholds%', 'not_us_citizen%', 'unemployed%',
       'vacant_housing_units%']]
```

```python
data = df.to_numpy()
X = data
```

```python
X.shape
```

```python
X_norm = X / X.std(axis=0) # Normalizing the feature columns
```

```python
features = df.columns
features
```

```python
figure, correlation_matrix = plot_pca_correlation_graph(X_norm, 
                                                        features,
                                                        dimensions=(1, 2, 3, 4, 5),
                                                        figure_axis_size=10)
```

## Consider smaller sets of features for better visibility

Unfortunatelly plot_pca_correlation_graph doesn't allow us to change font size, so it is very difficult to interpret the graph. We will build 4 circles with less features.

```python
featuresNestedList = []
featuresPerGraph = 5
numOfGraphs = len(features)//featuresPerGraph 

for i in range(numOfGraphs):
    if i == numOfGraphs-1:
        featuresNestedList.append(features[featuresPerGraph*i:])
    else:
        featuresNestedList.append(features[featuresPerGraph*i:featuresPerGraph*(i+1)])
featuresNestedList
```

```python
figures, correlation_matrices = [], []

for features in featuresNestedList:
    df_temp = df[features]
    data = df_temp.to_numpy()
    X = data
    X_norm = X/X.std(axis=0)
    figure, correlation_matrix = plot_pca_correlation_graph(X_norm, 
                                                        features,
                                                        dimensions=(1, 2, 3),
                                                        figure_axis_size=12)
    figures.append(figure)
    correlation_matrices.append(correlation_matrix)
    
```
